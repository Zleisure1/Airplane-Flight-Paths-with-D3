```python
import pandas as pd
```

```python
airport_list = "Resources/list-of-airports-in-united-states-of-america-hxl-tags-1.csv"
airport_list_df = pd.read_csv(airport_list)
```

```python
airport_list_df.head()
```

```python
airport_list_df = airport_list_df.drop(["continent", "iso_country", "iso_region", "municipality", "gps_code", "iata_code", "wikipedia_link", "keywords"], axis=1)
airport_list_df.head()
```

```python
airport_list_df = airport_list_df.drop(["ident"], axis=1)
airport_list_df.head()
```

```python
airport_list_df = airport_list_df.drop(["scheduled_service", "last_updated"], axis=1)
airport_list_df.head()
```

```python
airport_list_df = airport_list_df.drop([0])
```

```python
airport_list_df.head()
```

```python
airport_list_df.set_index('id')
airport_list_df.head()
```
