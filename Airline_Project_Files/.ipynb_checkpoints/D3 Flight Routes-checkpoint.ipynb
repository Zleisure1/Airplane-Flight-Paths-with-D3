```python
import pandas as pd
```

```python
pass_trips = pd.read_csv('MARKET_ALL_CARRIER.csv')
airports = pd.read_csv('airports.dat',names = ['Airport ID','Airport Name','Country','Name','IATA/FAA',                                            
                                               'ICAO','Latitude','Longitude','Altitude',
                                               'Timezone','DST','Tz database time zone','ICAO_2','IATA/FAA_2',])
```

```python
pass_trips.head()
```

```python
pass_trips = pass_trips[pass_trips['PASSENGERS'] > 0][['PASSENGERS','ORIGIN','DEST']]
print(len(pass_trips))
pass_trips.head()
```

```python
airports.head()
```

```python
airports = airports[['IATA/FAA','Latitude','Longitude']]
airports.head()
```

```python
pass_trips = pd.merge(pass_trips,airports,left_on='ORIGIN',right_on='IATA/FAA')
print(len(pass_trips))
pass_trips.head()
```

```python
pass_trips.columns
```

```python
pass_flights = pass_trips[['PASSENGERS','ORIGIN','DEST','Latitude','Longitude']]
pass_flights.head()
```

```python
pass_trips = pd.merge(pass_trips,airports,left_on='DEST',right_on='IATA/FAA')
print(pass_trips)
pass_trips.head()
```

```python
del pass_trips['IATA/FAA_x']
pass_trips.columns = ['PASSENGERS','ORIGIN','DEST',
                           'Latitude_x','Longitude_x',
                           'Latitude_y','Longitude_y']
pass_trips.head()
```

```python
pass_trips.sort_values(by='PASSENGERS',ascending=False,inplace=True)
pass_trips.head()
```

```python
total_pass_trips = sum(pass_trips['PASSENGERS'])
total_pass_trips
```

```python
pass_trips['proportion_of_pass_trips'] = pass_trips['PASSENGERS'] / total_pass_trips
pass_trips.head()
```

```python
pass_trips['cumsum_proportion'] = pass_trips['proportion_of_pass_trips'].cumsum()
pass_trips.head()
```

```python
pass_trips_sm = pass_trips[['Latitude_x','Longitude_x',
                                      'Latitude_y','Longitude_y',
                                      'cumsum_proportion']]
decimals = pd.Series([2, 2, 2, 2, 6], index=['Latitude_x','Longitude_x',
                                             'Latitude_y','Longitude_y',
                                             'cumsum_proportion'])
pass_trips_sm = pass_trips_sm.round(decimals)
pass_trips_sm.head()
```

```python
pass_trips_sm.to_csv('passenger_trips.csv',index=False)
```
